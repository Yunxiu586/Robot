```python
# map映射
# 将函数应用到可迭代对象的每个元素上，返回新的迭代器
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
print(squared)
```

```python
# filter过滤
# 筛选出满足函数条件的元素，返回新的迭代器
nums = [1, 2, 3, 4]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
from functools import reduce
# reduce化简
# 用函数对 可迭代对象 进行累积计算
nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)  # 1 * 2 * 3 * 4 = 24
```

### 遍历循环

```python
# while循环遍历列表
def while_traversal(list_traversal):
    """
    while循环遍历列表元素
    :param list_traversal: 遍历列表
    :return: None
    """
    ind=0 # 初始化
    while ind<len(list_traversal):
        print(list_traversal[ind])
        ind+=1

while_traversal([1,3,5,7])
```

```python
def for_traversal(list_traversal):
    """
    for循环遍历列表元素
    :param list_traversal: 遍历列表
    :return: None
    """
    for ele in list_traversal:
        print(ele)

for_traversal([0,2,4,6])
```

```python
# 遍历输出所有偶数
ind=0
num_list=list(range(21))
while ind<len(num_list):
    ele=num_list[ind]
    ind+=1
    if ele%2==0:
        print(f"{ele} ",end="")
    else:
        continue
```

```python
# 遍历输出所有偶数
for ele in num_list:
    if ele%2==0:
        print(f"{ele} ",end="")
    else:
        continue
```

```python
def word_frequency(text: str):
    text = text.lower()
    for char in text:
        if not char.isalnum() and char != ' ':
            # .isalnum()检查字符串是否只包含字母和数字字符
            text = text.replace(char, ' ')
    
    words = text.split()
    frequency_dict = {}
    
    for word in words:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    
    return frequency_dict
    
text = "Hello world! Hello NLP. World is fascinating."
frequency_dict = word_frequency(text)
frequency_dict
```

```python
def build_bow(corpus: list):
    """
    构建词袋模型Bag of Words
    参数:corpus: list - 由多个句子组成的文本列表
    返回:list - 包含词典和词频向量的列表
    """
    # 构建词典（按单词在语料中第一次出现的顺序）
    vocab = []
    for sentence in corpus:
        words = sentence.lower().split()
        for word in words:
            if word not in vocab:
                vocab.append(word)
    
    # 为每个句子生成词频向量
    vectors = []
    for sentence in corpus:
        # 初始化全0向量
        vector = [0] * len(vocab)
        words = sentence.lower().split()
        for word in words:
            # 找到单词在词典中的位置
            index = vocab.index(word)
            # 对应位置计数加1
            vector[index] += 1
        vectors.append(vector)
    
    return [vocab] + vectors

corpus = ["NLP is fun", "Learning NLP and deep learning", "Deep learning is the future"]
result = build_bow(corpus)
result
```

```python
'''
运算符 & ：位与运算；集合交集运算
运算符 | : 位或运算；集合并集运算
'''

def find_intersection_and_union(set1, set2):
    intersection = set1 & set2
    union = set1 | set2
    # intersection = set1.intersection(set2)
    # union = set1.union(set2)
    return intersection, union
    
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
intersection, union = find_intersection_and_union(set1, set2)
intersection, union
```

```python
print(f"字符串{my_str}转元组{tuple(my_str)}")
print(f"字典{my_dict}转元组{tuple(my_dict)}") # Key转换元组
print(f"字典{my_dict}转元组{tuple(my_dict.items())}")
```

```python
print(f"字符串{my_str}转集合{set(my_str)}") # 集合无序
print(f"字典{my_dict}转集合{set(my_dict)}") # Key转换集合
print(f"字典{my_dict}转集合{set(my_dict.items())}")
```

```python
print(f"列表{my_list}转字符串{str(my_list)}，数据类型是{type(str(my_list))}") # 打印结果形式相同，数据类型已经改变
print(f"元组{my_tuple}转字符串{str(my_tuple)}，数据类型是{type(str(my_tuple))}")
print(f"字典{my_dict}转字符串{str(my_dict)}") # 'Key'：Value
print(f"字典{my_dict}转字符串{str(my_dict.items())}")
```

```python
# 排序
# sorted()返回列表list
my_list=[1,3,2]
my_tuple=(1,3,2)
my_str="acebd"
my_set={1,3,2}
my_dict={"key1":1,"key3":2,"key2":3}
```

```python
# enumerate(iterable, start=0)
# iterable：一个可迭代对象,如列表、元组、字符串等
# start：索引的起始值，默认为0
fruits = ['apple', 'banana', 'cherry']
my_str="acebd"
my_dict = {"key1":1,"key3":2,"key2":3}
a = enumerate(fruits)
b = enumerate(my_dict)
a, b
```

```python
# 从 1 开始计数
for i, fruit in enumerate(fruits, start=1):
    print(f"第 {i} 个水果: {fruit}")
```

```python
# 创建 {索引: 元素} 的字典
index_dict = {i: fruit for i, fruit in enumerate(fruits)}
print(index_dict)  # 输出: {0: 'apple', 1: 'banana', 2: 'cherry'}
```

```python
# 查找 'banana' 的位置
for i, fruit in enumerate(fruits):
    if fruit == 'banana':
        print(f"香蕉的位置是: {i}")
        break
```

```python
# 迭代的坑
nums = [1, 2, 3, 4]
for idx, n in enumerate(nums):
    if n % 2 == 0:
        nums.remove(n)  # 跳过元素
        print(idx, nums)
```

```python
nums = [1, 2, 3, 4]
for idx, n in enumerate(nums[:]):  # 创建副本进行遍历
    if n % 2 == 0:
        nums.remove(n)  # 在原列表上删除
        print(idx, nums)
```

```python
nums = [1, 2, 3, 4]
for idx in range(len(nums)-1, -1, -1):  # 从后往前遍历
    n = nums[idx]
    if n % 2 == 0:
        removed_n = nums.pop(idx)  # 删除当前索引的元素
        print(idx, nums)
```
