```python
# 变量无类型，变量存储的数据有类型
# type()函数返回数据类型
str_type=type("A") # 字符串string
int_type=type(1) # 整数int
flo_type=type(1.1) # 浮点数float
print(str_type,type(1),flo_type)
```

```python
print(str(65), type(chr(65)))
print(chr(65), type(chr(65))) # chr()将整数转换为对应的字符
print(ord('A'), type(ord('A'))) # ord()是chr()的逆操作
print('A'.encode(), type('A'.encode())) # 字符串转换为字节
print(b'A'.decode(), type(b'A'.decode()))
```

```python
print(int(1.5)) # 丢失精度,无四舍五入
int("1") # 纯数字字符串才能准换为数字
int('123',16) # 字符串 '123' 按照十六进制(16进制)解析为十进制整数
```

```python
float(1)
print(float("1.5"),type(float("1.5")))
str(1) # 任意类型数据都可以转换为字符串
```

```python
# 字符串拼接
year = "2025" # 字符型变量
month = "6"
day = "21"
print("今天是" + year + "年" + month + "月" + day + "日")
```

```python
# 字符串格式化
year = "2025"
month = 6 # 数字类型
day = 21
time = "今天是%s年%s月%s日" %(year,month,day)
# %表示占位，s占位符：将变量变为字符串放入占位
print(time)
```

```python
name = "我"
age = 20
height = 1.85
message = "%s今年%4d岁，身高%5.2f米"%(name,age,height)
# %d整数，%f浮点数
print(message)
message = "%s今年%4d岁，身高%5.2f米"%(name,age,height)
# m.n m控制宽度，宽度小于数字本身不生效，宽度大于数字在左侧插入空格；n小数精度，会四舍五入
print(message)
```

```python
print(f"{name}今年{age}岁，身高{height}米")
# 快速格式化，原样输出，不理会类型，不控制精度 f"内容{变量}"
```

```python
# 表达式的格式化
print("2*2=%d"%(2*2))
print(f"2*2={2*2}")
print("字符串在Python中的类型名是：%s" % type("字符串"))
```
