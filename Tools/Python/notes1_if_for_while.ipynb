```python
# if-elif-else
# 判断是互斥的，顺序执行，仅执行第一个满足条件的语句
height = 130
vip_level = 5
day = 1
if height < 120:
    print("身高小于120cm，免费游玩")
elif vip_level > 3:
    print("vip等级大于3，免费游玩")
elif day == 1:
    print("免费日，免费游玩")
else:
    print("票价10元")
```

```python
# 判断语句的嵌套
# 通过缩进决定多层次嵌套关系
age = -1
if age > 0:
    if age >= 18:
        print("已成年")
    else:
        print("未成年")
else:
    print("ERROR")
```

```python
# 打印九九乘法表
i = 1 # 行
while i < 10:
    j = 1 # 列
    while j <= i:
        print(f"{j}*{i}={i*j}\t",end="") # \t 制表符对齐 ,end="" print不换行
        j += 1
    i += 1
    print() # 换行
```

```python
# for循环 遍历循环 轮询机制
# for 临时变量 in 序列：
# 临时变量的作用域在for循环外部
message="Digital Signal Process"
cnt=0
for x in message:
    if x == 'a':
        cnt+=1

print(cnt)
```

```python
def newton_sqrt(S, tolerance=1e-4, max_iterations=100):
    x_n = S / 2  # 初始猜测
    for _ in range(max_iterations):
        x_next = 0.5 * (x_n + S / x_n)  # 更新迭代值
        if abs(x_next - x_n) < tolerance:  # 检查收敛性
            break
        x_n = x_next
    return x_n

S = 2
sqrt_S = newton_sqrt(S)
print(f"{S} 的平方根是: {sqrt_S:.25f}")
```
