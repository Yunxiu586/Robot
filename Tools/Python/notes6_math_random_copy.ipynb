# math

```python
import math

# 数值常数
print(f"数学常数 e: {math.e}")
print(f"圆周率 π: {math.pi}")
print(f"无穷大 inf: {math.inf}")
print(f"非数字 nan: {math.nan}")
```

```python
# 数值运算函数
x = 2.25
y = 4.9
print(f"ceil({x}) = {math.ceil(x)}")  # 向上取整
print(f"floor({x}) = {math.floor(x)}")  # 向下取整
print(f"trunc({x}) = {math.trunc(x)}")  # 截断小数部分
print(f"fabs({x}) = {math.fabs(x)}")  # 绝对值
print(f"sqrt({x}) = {math.sqrt(x)}")  # 平方根
print(f"pow({x}, {y}) = {math.pow(x, y)}")  # 幂运算
print(f"exp(2) = {math.exp(2)}")  # e的x次幂
print(f"ln(e) = {math.log(math.e)}")  # 自然对数
print(f"log10(100) = {math.log10(100)}")  # 以10为底的对数
```

```python
# 三角函数 (使用弧度制)
angle_deg = 45
angle_rad = math.radians(angle_deg)  # 角度转弧度
angle_rad = angle_deg * math.pi / 180
print(f"{angle_deg}° = {angle_rad} 弧度")
print(f"sin({angle_rad}) = {math.sin(angle_rad)}")
print(f"cos({angle_rad}) = {math.cos(angle_rad)}")
print(f"tan({angle_rad}) = {math.tan(angle_rad)}")
```

```python
# 其他函数
print(f"gcd(12, 18) = {math.gcd(12, 18)}")  # 最大公约数
print(f"factorial(5) = {math.factorial(5)}")  # 阶乘
```

# random

```python
import random
# 生成[0.0, 1.0)之间的随机浮点数
random.random()
```

```python
# uniform(a, b) - 生成[a, b]之间的随机浮点数
random.uniform(5, 10)
```

```python
# randint(a, b) - 生成[a, b]之间的随机整数
random.randint(1, 6)
```

```python
# choice(seq) - 从序列中随机选择一个元素
fruits = ['苹果', '香蕉', '橙子', '葡萄', '草莓']
random.choice(fruits)
```

```python
# shuffle(seq) - 将序列随机排序
cards = ['红桃A', '黑桃K', '方块Q', '梅花J', '红桃10']
random.shuffle(cards)
print(cards)
```

```python
# sample(population, k) - 从总体中随机选择k个不重复的元素
numbers = list(range(1, 11))
selected = random.sample(numbers, 3)
print(selected)
```

```python
# randrange(start, stop, step)从指定范围随机选择一个数
random.randrange(0, 11, 2)
```

```python
# 设置随机数种子
random.seed(42)
print(random.randint(1, 100), random.randint(1, 100))
random.seed(42)
print(random.randint(1, 100), random.randint(1, 100))
```

# copy

```python
import copy
# 原始列表
a = [1, 2]
b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝
# 修改拷贝后的列表
a[1] = [1, 2]
print("原列表:", a)
print("浅拷贝:", b)
print("深拷贝:", c)
```

```python
a = [3, 4]
b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝
b[1] = [3, 4]
print("原列表:", a)
print("浅拷贝:", b) 
print("深拷贝:", c)
```

```python
a = [5, 6]
b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝
c[1] = [5, 6]
print("原列表:", a)
print("浅拷贝:", b) 
print("深拷贝:", c)
```

```python
a = [1, [0, 0]]
b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝
a[1][1] = 2           # 修改原列表嵌套对象，影响浅拷贝
print("原列表:", a)
print("浅拷贝:", b) 
print("深拷贝:", c)
```

```python
a = [3, [0, 0]]
b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝
b[1][1] = [4, 4]      # 修改浅拷贝嵌套对象，影响原列表
print("原列表:", a)
print("浅拷贝:", b) 
print("深拷贝:", c)
```

```python
a = [5, [0, 0]]
b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝
c[1][1] = [6, 6]
print("原列表:", a)
print("浅拷贝:", b) 
print("深拷贝:", c)
```
