```python
def minus(x,y):
    """
    minus函数可以接收2个参数，进行2数相减
    :param x: 被减数
    :param y: 减数
    :return: 差
    """
    # 函数的说明文档，鼠标停留在函数上可以查看函数的说明文档
    z=x-y
    return z
print(f"5-2={minus(5,2)}")
```

```python
def hi():
    print("Hello World")
result=hi()
print(f"无返回值的函数，返回内容是{result}，类型是{type(result)}")
# None 字面量空
# None 用于函数返回值、布尔运算、定义变量
while not result:
    print("None在布尔运算中等于False")
    break
```

```python
def factor(n):
    """
    factor函数实现n的阶乘，即n!
    :param n: 正整数n
    :return: n！
    """
    if n >= 2:
        factor_n = n * factor(n-1)
        # 嵌套调用
    else:
        factor_n = 1
    return factor_n
print(factor(5))
```

```python
# 函数体内的变量是局部变量，生命周期随函数体结束而结束
num=2
def local_eg():
    num=1
    print(num)
    
def global_eg():
    global num
    num=3 # 全局变量
    print(num)

local_eg()
global_eg()
print(num)
```

```python
def muti_return(x,y,z):
    f1=x+y+z
    f2="xyz"
    f3=x>y
    return f1,f2,f3 # 多返回值，数据类型任意

a0,b0,c0=muti_return(1,2,3)
print(a0,b0,c0)

a1=muti_return(1,2,3)[0] # 下标索引
print(a1)
```

```python
# 位置参数
# 传递的参数和定义的参数的顺序和个数必须一致
def user_info1(name,age,gender):
    print(f"您的名字是{name},年龄是{age},性别是{gender}")

user_info1('Alice',20,'female')
```

```python
# 关键字参数
# Key = Value
# 位置参数必须在关键字参数的前面
# 关键字参数无先后顺序
user_info1('Liz',age=21,gender='female')
```

```python
# 缺省参数
def user_info2(name,age,gender='male'): # 默认参数必须在最后
    print(f"您的名字是{name},年龄是{age},性别是{gender}")

user_info2("Tim",22)
user_info2("Emmy",20,"female") # 默认参数允许覆盖
```

```python
# 不定长参数
# 位置传递不定长
def uncertain1(*args): # args是元组(tuple)类型
    return args

print(uncertain1("Mess"), type(uncertain1("Mess")))
print(uncertain1("Mess",35), type(uncertain1("Mess",35)))
```

```python
# 关键字传递不定长
def uncertain2(**kwargs): # key-word args是字典(dict)类型
    return kwargs

# 参数是 Key=Value 的形式
print(uncertain2(name='Mess',age=35,gender='male'), type(uncertain2(name='Mess',age=35,gender='male'))) 
```
