```python
class Directory2:

    def __init__(self,name,age,tel):
        self.name=name
        self.age=age
        self.tel=tel
        print(f'{self.name}年龄为{self.age},TA的电话号码是{self.tel}')

    # __str__字符串方法
    def __str__(self):
        return f'Directory2类对象是，name：{self.name},age:{self.age},tel:{self.tel}'

    # __lt__小于号比较方法
    def __lt__(self,other): # print(linkman2>=linkman3)不允许>=
        return self.age<other.age

    # __le__小于等于号比较方法
    def __le__(self,other): # print(linkman2>linkman3)不允许>
        return self.age<=other.age

    # __eq__等于号比较方法
    def __eq__(self,other):
        return self.age==other.age
```

```python
linkman2=Directory2('工藤新一',18,987654321)
```

```python
print(linkman2)
```

```python
print(str(linkman2))
```

```python
linkman3=Directory2('江户川柯南',8,24680246)
```

```python
'''
面向对象的三大特性
封装、继承、多态

1.封装
将现实世界事物的属性、行为封装到类中，描述为成员变量、成员方法

私有成员
在类中提供仅供内部使用的属性和方法，类对象无法使用
''';
```

"_"的作用
* _name 标定的内部函数（私有函数），不会被from XXX import * 调用 
* name_ 标定特殊函数名，不会和python关键词冲突
* \_\_name 私有类属性
* \_\_name\_\_ 预留函数，不要自己定义 (\_\_version\_\_)

```python
class Phone1:

    __current_voltage=1.5 # 私有成员变量
    producer = 'iPhone'

    def __keep_sigel_core(self): # 私有成员方法
        print('Phone1单核运行CPU')

    def call_by_5G(self):
        if self.__current_voltage>=1: # 访问私有成员变量
            print('Phone1的5G通话已开启')
        else:
            self.__keep_sigel_core() # 访问私有成员方法
            print('Phone1的电量不足，无法使用5G通话，已设置为单核运行')
```

```python
phone1=Phone1()
phone1.call_by_5G()
```

```python
# 访问私有变量（需要使用名称改写机制）
try:
    print(Phone1.__current_voltage)  # 报错
except Exception as e:
    print(e)
```

```python
print(Phone1._Phone1__current_voltage)
```

```python
Phone1._Phone1__current_voltage = 2
print(Phone1._Phone1__current_voltage)
```

```python
class Parent:
    _num = 10 # 可以继承访问
    def print_num(self):
        print(self._num)

class Child(Parent):
    def print_num(self):
        print(self._num)

p = Parent()
c = Child()

p.print_num()  # 10
c.print_num()  # 10
```

```python
class Parent:
    __num = 10
    def print_num(self):
        print(f"父类{self.__num}")

class Child(Parent):
    def print_num(self):
        try:
            print(self.__num)
        except Exception as e:
            self.__num = 20
            print(e)
            print(f"子类{self.__num}")

p = Parent()
c = Child()

p.print_num()  # 10
c.print_num()  # 20
```

```python
# 复写父类的属性和方法
class Phone3(Phone1):
    producer='HUAWEI'     # 复写父类属性

    def call_by_5G(self): # 复写父类方法
        print('Phone3开启CPU单核模式')
        print('Phone3使用5G网络通话')

        # 调用父类成员属性
        print(f'调用父类成员{Phone1.producer}')
        print(f'调用父类成员{super().producer}')
        # 调用父类成员方法
        Phone1.call_by_5G(self) # self
        super().call_by_5G()    # 空括号
```

```python
phone3=Phone3()
```

```python
print(f'phone1的生产厂商是{phone1.producer}') 
```

```python
print(f'phone3的生产厂商是{phone3.producer}') # 调用复写后的新属性
```

```python
phone3.call_by_5G()  # 调用复写后的新方法
```

```python
'''
3.多态
同样的行为(函数)，传入不同的对象，得到不同的状态
函数(方法)形参声明接收父类对象
实际传入父类对象的子类进行工作
父类定义声明。子类实际工作

父类确定有哪些方法，方法的实现由子类决定
抽象类(接口)：含有抽象方法的类
抽象方法：方法体是空(pass)
多用于顶层设计，便于子类具体实现
对子类的软性约束，要求子类必须复写(实现)父类的一些方法
''';
```

```python
class Animal:
    def speak(self):
        """所有动物都能发出声音"""
        pass

    def move(self):
        """所以动物都能移动"""
        pass
```

```python
class Dog(Animal):
    def speak(self):
        print('汪汪汪')

    def move(self):
        print('跑步')
```

```python
class Bird(Animal):
    def speak(self):
        print('啾啾啾')

    def move(self):
        print('飞翔')
```

```python
def make_noise(animal:Animal):
    animal.speak()
```

```python
dog=Dog()
bird=Bird()
```

```python
make_noise(dog)
make_noise(bird)
```

```python
dog.move()
bird.move()
```
