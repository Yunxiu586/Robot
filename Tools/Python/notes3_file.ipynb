```python
'''
文件的编码
UTF-8以3个字节表示中文，GBK以2个字节表示中文
\\uhhhh是 Unicode转义序列，表示一个Unicode字符，hhhh是一个四位十六进制数

1.文件的读取
打开文件
open(name,mode,encoding)
name：要打开的文件名的字符串(可包含文件所在具体路径)
mode：设置打开文件的模式(访问模式)：只读(r)、写入(w)、追加(a)等
encoding：编码格式(一般UTF-8)

r mode：默认模式
w mode:若文件已存在，则打开文件并从头编辑，原内容被删除，若文件不存在，创建新文件
a mode：若文件已存在，则新内容被写入已有内容之后，若文件不存在，创建新文件进行写入
''';
```

```python
file=open('D:/Python_study/PythonProject/demo1.txt','r',encoding='UTF-8') # /
print(type(file))
# encoding不是第三个参数，用关键字参数指定
```

```python
# read方法
print(f"读取6个字节的内容：\n{file.read(6)}") # 文件指针停留在第6个字节后面
print(f"读取剩余全部内容：\n{file.read()}")
```

```python
# seek方法&tell方法
file.seek(2,0)
# cookie：移动的字节数
# whence=0：默认、开头；whence=1：当前位置；whence=2末尾
print(f"指针所在位置是第{file.tell()}个字节")
```

```python
file.seek(0)
# readlines方法
lines=file.readlines() # 读取文件全部行，封装到列表中
print(f"lines对象的内容是{lines}") # \n换行
print(f"lines对象的类型是{type(lines)}") # list
```

```python
# readline方法
file.seek(0)
line1=file.readline()
line2=file.readline()
print(f"第一行：{line1}") # 换行符
print(f"第二行：{line2}")
```

```python
# 2.文件的关闭
file.close()
```

```python
# with open
# with open语句块中语句对文件进行操作，操作完成自动close
with open('./demo1.txt','r',encoding='UTF-8') as file: # 当前目录下的文件
    for line in file: # for循环每次记录一行数据
        print(f"{line}数据类型是{type(line)}") # str
```

```python
with open('./word.txt','r',encoding='UTF-8') as file:
    content=file.read()
    cnt=content.count("go")
    print(f"go出现的次数是{cnt}")
```

```python
with open('./word.txt','r',encoding='UTF-8') as file:
    cnt=0
    for lines in file:
        lines=lines.strip() # 去除换行符
        words=lines.split(" ")
        for word in words:
            if word=="go":
                cnt+=1
    print(f"go出现的次数是{cnt}")
```

```python
# 捕获异常
"""
try
    可能发生错误的代码
except
    如果出现异常执行的代码
""";
```

```python
try:
    a=1/0
    print(a,type(a))
except:
    a=float('inf')
    print(a,type(a))
```

```python
# 捕获指定异常
try:
    print(name)
except NameError as e:
    print("变量未定义")
    print(e,type(e))
```

```python
# 捕获多个异常
try:
    print(name)
    a=1/0
except (NameError,ZeroDivisionError) as e:
    print("变量未定义 或 除0")
    print(e,type(e))
```

```python
# 捕获全部异常
try:
    file=open('D:/Python_study/PythonProject/abcd.txt','r',encoding='UTF-8')
except Exception as e:
    print(f"出现异常,异常类型是{e}")
```

```python
# else&finally
try:
    print("Hello World")
except Exception as e:
    print(f"出现异常,异常类型是{e}")
else: # 无异常执行
    print("无异常")
finally: # 一定执行
    print("无论是否有异常，都执行")
```

```python
# 异常具有传递性
def func1():
    a=1/0

def func2():
    try:
        func1()
    except Exception as e:
        print(f"调用的函数存在异常,异常的类型是{type(e)}")

func2()
```
