广播机制(Broadcasting Mechanism)

维度对齐：从最右边的维度开始向左比较
维度兼容：每个维度必须满足其一
维度大小相等
其中一该维度大小为1，则可以广播到另一个数组的大小
其中张量数组缺少该维度，可以添加大小为1的维度

```python
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a,b
```

```python
a + b # 3 by 2
```

```python
X[0:2,:] = 12
X
```

```python
before = id(Y) # 内存地址
Y = Y + X
id(Y) == before # 分配新地址
```

```python
# 执行原地操作:不分配新内存
Z = torch.zeros_like(Y)
print('id(Z)',id(Z))
Z[:] = X + Y
print('id(Z)',id(Z))
```

```python
before = id(Y) 
Y += X
id(Y) == before # 减少内存开销
```

```python
before = id(Y) 
Y[:] =  Y + X
id(Y) == before # 
```

```python
a = torch.tensor([1, 2, 3, 4])
b = torch.zeros_like(a)
b = a.clone() # 分配新内存，将 a 的一个副本分配给 b
b[0] = 10 # b 改变不影响 a
a, b
```

```python
a = torch.tensor([1, 2, 3, 4])
b = torch.zeros_like(a)
b_copied = b.copy_(a) # 复制
b[0] = 10 # 修改 b 不影响 a ,影响 b_copied
a, b, b_copied
```

```python
a = torch.tensor([1, 2, 3, 4])
b = a.expand(2,4)
b[0][0] = 10 # 修改 b 影响 a
a, b
```

```python
a = torch.tensor([1, 2, 3, 4])
b = a[1:5]   # 切片
b[0] = 10    # 修改 b 影响 a
a, b
```

```python
a = torch.tensor([1, 2, 3, 4])
b = a.view(torch.numel(a)) # 视图
b[0] = 10    # 修改 b 影响 a
a, b
```

```python
a = torch.tensor([1, 2, 3, 4])
b = a.add_(1) # 原地加 1，修改 b 影响 a
a, b
```

```python
a = torch.tensor([1, 2, 3, 4])
b = a.mul_(2) # 原地乘 2，修改 b 影响 a
a, b
```

# 数据预处理

```python
import os
```

创建一个人工数据集，存储在.csv文件
CSV(Comma-Separated Values)文件，中文叫逗号分隔值文件，是一种非常简单、通用的纯文本格式，用来存储表格数据

逗号分隔不同的数据字，,换行符来分隔每一行数据

```python
os.makedirs(os.path.join('D:/Python_study/DL_1','houseData'),exist_ok=True) # 创建目录
data_file = os.path.join('D:/Python_study/DL_1','houseData','house_tiny.csv')
with open(data_file,'w') as f:
    f.write('NumRooms,Ally,Price\n') # 列名
    f.write('NA,Pave,127500\n') # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

```python
import pandas as pd
```

```python
data = pd.read_csv(data_file)
data
```

```python
inputs, outputs = data.iloc[:,0:2], data.iloc[:,2]
numeric_cols = inputs.select_dtypes(include=['number']).columns
inputs[numeric_cols] = inputs[numeric_cols].fillna(inputs[numeric_cols].mean())
print(inputs)
```

```python
inputs = pd.get_dummies(inputs,dummy_na=True) # 独热编码，分类变量
# dummy_na=True 缺失值创建一个单独的虚拟列
print(inputs)
```

```python
type(inputs.values)
```

```python
inputs = inputs.astype(float)
outputs = outputs.astype(float)
X, y = torch.tensor(inputs.values),torch.tensor(outputs.values)
X, y
```
